# ML Model

```python
x=df.drop(['salary','salary_in_usd'],axis=1)
```

```python
y=df['salary_in_usd']
```

```python
sns.boxplot(y)
```

```python
cat_col=df.select_dtypes(include='object').columns.tolist()
```

```python
cat_col
```

```python
df2=pd.get_dummies(df,dtype=float)
```

```python
df2
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df2,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.2f}")

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse:.2f}")

# R-squared (Coefficient of Determination)
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
```

# Our linear regression model is showing good accuracy but this can be a result of small dataset also i.e, may be model does not predict well for new data
